### Amount of Times Someone Brushes Their Teeth (Per Day)

```python
xCol = dataFrame['How many times per day do you brush your teeth?']
x = xCol.tolist()
x
```

```python
import collections
counts = collections.Counter(x)

# Generate a list of numbers from 0 to 10 and their respective counts
numbers_range = list(range(11))
counts_list = [counts[num] for num in numbers_range]
counts_list
```

#### 5 Number Summary

```python
import seaborn as sns
series = pd.Series(x)
mean = series.mean()
median = series.median()
mode = series.mode().values[0]

summary_df = pd.DataFrame({
    'Statistic': ['Mean', 'Median', 'Mode'],
    'Value': [mean, median, mode]
})
summary_df
```

### What type of toothbrush do you currently use?

```python
xCol2 = dataFrame['What type of toothbrush do you currently use?']
x2 = xCol2.tolist()
x2
```

```python
counts2 = collections.Counter(x2)
types_to_count = ["Regular Toothbrush", "Electric Toothbrush", "Other"]

# Generate a list of numbers from 0 to 10 and their respective counts
counts_list2 = [counts2[type] for type in types_to_count]
counts_list2
```

## Summarization of Question #1


```python
import numpy as np

Q1 = np.percentile(x, 25)
Q3 = np.percentile(x, 75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print("The IQR for question #1 is: ", str(IQR))
print("The lower bound for question #1 is: ", str(lower_bound))
print("The upper bound for question #1 is: ", str(upper_bound))
```

**Question number one states: "How many times do you brush your teeth? (Per Day)" With the given sample population of n = 44 the data has been organized in a visual manor to display the following points.**



*   There is indeed an outlier in this sample population which is 8 times. Due to some people either joking or for some reason seriously brushing 8 times a day, they exceed the upper bound being 2.0 for this sample population and therefore are outliers or in other words are an incredibely niche set of people compared to most (and are most likely joking if not you need help).
*   On average most people brush around 2.25 times a day
*   The most common amount times people brush is 2 times a day.
*   The median is 2.0 times per day.
*   We can conclude from this data that the majority of people brush their teeth approximately 2 times a day with an outlier of 8 portraying either crazy people or people who are joking.
